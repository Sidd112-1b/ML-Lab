```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
df=pd.read_excel(r"C:\Users\Siddharth\Downloads\Lab Session Data.xlsx",sheet_name="IRCTC Stock Price")
df["Date"] = pd.to_datetime(df["Date"])
price = df["Price"]
```

```python
numpy_mean = np.mean(price)
numpy_variance = np.var(price)

print("Using NumPy")
print("Mean =", numpy_mean)
print("Variance =", numpy_variance)
```

```python
def my_mean(data):
    total = 0

    for x in data:
        total += x

    return total / len(data)


def my_variance(data):
    mean = my_mean(data)

    total = 0

    for x in data:
        total += (x - mean) ** 2

    return total / len(data)
```

```python
custom_mean = my_mean(price)
custom_variance = my_variance(price)

print("\nUsing User-defined Functions")
print("Mean =", custom_mean)
print("Variance =", custom_variance)
```

```python
numpy_times = []
custom_times = []

for i in range(10):

    start = time.perf_counter()

    np.mean(price)
    np.var(price)

    end = time.perf_counter()

    numpy_times.append(end - start)


for i in range(10):

    start = time.perf_counter()

    my_mean(price)
    my_variance(price)

    end = time.perf_counter()

    custom_times.append(end - start)


print("\nAverage NumPy Time =", np.mean(numpy_times))
print("Average Custom Function Time =", np.mean(custom_times))
```

```python
wednesday = df[df["Date"].dt.day_name() == "Wednesday"]

wednesday_mean = wednesday["Price"].mean()

print("\nPopulation Mean =", numpy_mean)
print("Wednesday Mean =", wednesday_mean)
```

```python
april = df[df["Date"].dt.month == 4]

april_mean = april["Price"].mean()

print("\nApril Mean =", april_mean)
```

```python
loss = df[df["Chg%"].apply(lambda x: x < 0)]

prob_loss = len(loss) / len(df)

print("\nProbability of Loss =", prob_loss)
```

```python
profit_wednesday = wednesday[wednesday["Chg%"] > 0]

prob_profit_wednesday = len(profit_wednesday) / len(df)

print("\nProbability of Profit on Wednesday =", prob_profit_wednesday)
```

```python
conditional_probability = len(profit_wednesday) / len(wednesday)

print("Conditional Probability P(Profit | Wednesday) =",
      conditional_probability)
```
