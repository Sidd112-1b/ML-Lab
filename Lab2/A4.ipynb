```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"C:\Users\Siddharth\Downloads\Lab Session Data.xlsx",sheet_name="thyroid0387_UCI")
print("First 5 rows:\n")
print(df.head())

print("\n")
print("Column Names")
print(df.columns.tolist())

print("\n")
print("Data Types")
print(df.dtypes)
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

categorical_cols = df.select_dtypes(exclude=np.number).columns

print("\n")
print("Numerical Columns")
print(list(numeric_cols))

print("\n")
print("Categorical Columns")
print(list(categorical_cols))
```

```python
print("\n")
print("Categorical Attributes and Suggested Encoding")

for col in categorical_cols:

    unique_values = df[col].unique()

    print("\nAttribute :", col)
    print("Unique Values :", unique_values)

    if len(unique_values) <= 2:
        print("Suggested Encoding : Label Encoding (Binary)")
    else:
        print("Suggested Encoding : One-Hot Encoding")
```

```python
print("\n")
print("Range of Numerical Variables")

for col in numeric_cols:

    minimum = df[col].min()
    maximum = df[col].max()

    print(f"{col}")
    print("Minimum =", minimum)
    print("Maximum =", maximum)
    print("Range =", maximum - minimum)
    print()
```

```python
print("\n")
print("Missing Values")

print(df.isnull().sum())
```

```python
print("\n")
print("Outliers")

for col in numeric_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col} : {len(outliers)} outliers")
```

```python
print("\n")
print("Mean, Variance and Standard Deviation")

for col in numeric_cols:

    print(f"\n{col}")

    print("Mean =", df[col].mean())

    print("Variance =", df[col].var())

    print("Standard Deviation =", df[col].std())
```
