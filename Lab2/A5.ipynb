```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"C:\Users\Siddharth\Downloads\Lab Session Data.xlsx",sheet_name="thyroid0387_UCI")
```

```python
binary_columns = []

for col in df.columns:

    unique_values = df[col].dropna().unique()

    if len(unique_values) == 2:
        binary_columns.append(col)

print("Binary Attributes:")
print(binary_columns)
```

```python
v1 = df.loc[0, binary_columns]
v2 = df.loc[1, binary_columns]
```

```python
mapping = {
    't': 1, 'f': 0,
    'T': 1, 'F': 0,
    'yes': 1, 'no': 0,
    'Yes': 1, 'No': 0,
    True: 1, False: 0
}

v1 = v1.replace(mapping)
v2 = v2.replace(mapping)

v1 = pd.to_numeric(v1)
v2 = pd.to_numeric(v2)
```

```python
f11 = 0
f10 = 0
f01 = 0
f00 = 0

for i in range(len(v1)):

    if v1.iloc[i] == 1 and v2.iloc[i] == 1:
        f11 += 1

    elif v1.iloc[i] == 1 and v2.iloc[i] == 0:
        f10 += 1

    elif v1.iloc[i] == 0 and v2.iloc[i] == 1:
        f01 += 1

    elif v1.iloc[i] == 0 and v2.iloc[i] == 0:
        f00 += 1

print("\nf11 =", f11)
print("f10 =", f10)
print("f01 =", f01)
print("f00 =", f00)
```

```python
JC = f11 / (f01 + f10 + f11)

SMC = (f11 + f00) / (f00 + f01 + f10 + f11)

print("\nJaccard Coefficient =", JC)
print("Simple Matching Coefficient =", SMC)
```
