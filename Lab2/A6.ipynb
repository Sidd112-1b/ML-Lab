```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"C:\Users\Siddharth\Downloads\Lab Session Data.xlsx",sheet_name="thyroid0387_UCI")
mapping = {
    't': 1, 'f': 0,
    'T': 1, 'F': 0,
    'yes': 1, 'no': 0,
    'Yes': 1, 'No': 0,
    True: 1, False: 0
}

df = df.replace(mapping)
```

```python
numeric_df = df.select_dtypes(include=np.number)

A = numeric_df.iloc[0].to_numpy()
B = numeric_df.iloc[1].to_numpy()

print("Vector A:")
print(A)

print("\nVector B:")
print(B)
```

```python
dot_product = np.dot(A, B)

norm_A = np.linalg.norm(A)
norm_B = np.linalg.norm(B)

cosine_similarity = dot_product / (norm_A * norm_B)

print("\nDot Product =", dot_product)
print("||A|| =", norm_A)
print("||B|| =", norm_B)
print("\nCosine Similarity =", cosine_similarity)
```
