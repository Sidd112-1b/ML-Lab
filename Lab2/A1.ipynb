```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\Siddharth\Downloads\Lab Session Data.xlsx",sheet_name="Purchase data")
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
print(df.head())
print(df.columns.tolist())
X=df[["Candies (#)","Mangoes (Kg)","Milk Packets (#)"]]
y=df[["Payment (Rs)"]]

print("\n Feature Matrix X: ")
print(X)

print("\n Output Matrix y: ")
print(y)
```

```python
rank=np.linalg.matrix_rank(X)
print("Rank of feature matrix: ", rank)
print("\n")
psinv=np.linalg.pinv(y)
print("Cost of each product:", psinv)
```
